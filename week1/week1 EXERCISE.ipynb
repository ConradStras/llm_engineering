```python
# imports

import os
import json
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from scraper import fetch_website_links, fetch_website_contents
from openai import OpenAI
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up env
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

MODEL = 'gpt-4o-mini'
openai = OpenAI()
```

```python
# here is the question; type over this to ask something new

user_prompt = """ 
Explain or clarify the following to me:

"""
system_prompt = """
    You are a helpful tutor that is able provide help on a wide array of topics. 
    You do this by explaining the subject matter in a step by step manner.
    Your response usually includes a practical example of the topic at hand.
    Respond in markdown, do not include your markdown in a code block, just provide the markdown.
"""  

def generateResponse(topic):
    stream = openai.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content":user_prompt+topic}],
            stream=True
    ) 
    result = stream
    return result
```

```python
# Get gpt-4o-mini to answer, with streaming
def stream_response(topic):
    stream = generateResponse(topic)
    display_handle = display(Markdown(""), display_id=True)
    response=""
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response), display_id=display_handle.display_id)
stream_response("Getting started with C++ and C in the agentic coding era")
```
